```python
print("Installing packages...")
!pip install -q ipywidgets
from google.colab import files
import numpy as np
import matplotlib.pyplot as plt
from tensorflow import keras
from tensorflow.keras import layers
from PIL import Image, ImageOps
import io
from IPython.display import display, HTML, clear_output
import ipywidgets as widgets

print("Packages installed successfully!")

class NeuralNetworkScratch:

    def __init__(self, layer_sizes):
        self.layer_sizes = layer_sizes
        self.weights = []
        self.biases = []

        for i in range(len(layer_sizes) - 1):
            w = np.random.randn(layer_sizes[i], layer_sizes[i+1]) * np.sqrt(2.0/layer_sizes[i])
            b = np.zeros((1, layer_sizes[i+1]))
            self.weights.append(w)
            self.biases.append(b)

    def sigmoid(self, z):
        return 1 / (1 + np.exp(-np.clip(z, -500, 500)))

    def sigmoid_derivative(self, z):
        s = self.sigmoid(z)
        return s * (1 - s)

    def relu(self, z):
        return np.maximum(0, z)

    def relu_derivative(self, z):
        return (z > 0).astype(float)

    def softmax(self, z):
        exp_z = np.exp(z - np.max(z, axis=1, keepdims=True))
        return exp_z / np.sum(exp_z, axis=1, keepdims=True)

    def forward_propagation(self, X):
        self.activations = [X]
        self.z_values = []

        for i in range(len(self.weights) - 1):
            z = np.dot(self.activations[-1], self.weights[i]) + self.biases[i]
            self.z_values.append(z)
            a = self.relu(z)
            self.activations.append(a)

        z = np.dot(self.activations[-1], self.weights[-1]) + self.biases[-1]
        self.z_values.append(z)
        a = self.softmax(z)
        self.activations.append(a)

        return self.activations[-1]

    def backward_propagation(self, X, y, learning_rate):
        m = X.shape[0]

        dz = self.activations[-1] - y
        dw = np.dot(self.activations[-2].T, dz) / m
        db = np.sum(dz, axis=0, keepdims=True) / m

        gradients_w = [dw]
        gradients_b = [db]

        for i in range(len(self.weights) - 2, -1, -1):
            dz = np.dot(dz, self.weights[i+1].T) * self.relu_derivative(self.z_values[i])
            dw = np.dot(self.activations[i].T, dz) / m
            db = np.sum(dz, axis=0, keepdims=True) / m

            gradients_w.insert(0, dw)
            gradients_b.insert(0, db)

        for i in range(len(self.weights)):
            self.weights[i] -= learning_rate * gradients_w[i]
            self.biases[i] -= learning_rate * gradients_b[i]

    def train(self, X, y, X_val, y_val, epochs, learning_rate, batch_size=128):
        train_losses = []
        train_accuracies = []
        val_losses = []
        val_accuracies = []

        print(f"Training Neural Network from Scratch...")
        print(f"Architecture: {' -> '.join(map(str, self.layer_sizes))}")
        print(f"Parameters: {sum(w.size for w in self.weights):,}")
        print("-" * 60)

        for epoch in range(epochs):
            indices = np.random.permutation(X.shape[0])
            X_shuffled = X[indices]
            y_shuffled = y[indices]

            for i in range(0, X.shape[0], batch_size):
                X_batch = X_shuffled[i:i+batch_size]
                y_batch = y_shuffled[i:i+batch_size]

                self.forward_propagation(X_batch)
                self.backward_propagation(X_batch, y_batch, learning_rate)

            train_pred = self.forward_propagation(X)
            train_loss = -np.mean(np.sum(y * np.log(train_pred + 1e-8), axis=1))
            train_acc = np.mean(np.argmax(train_pred, axis=1) == np.argmax(y, axis=1))

            val_pred = self.forward_propagation(X_val)
            val_loss = -np.mean(np.sum(y_val * np.log(val_pred + 1e-8), axis=1))
            val_acc = np.mean(np.argmax(val_pred, axis=1) == np.argmax(y_val, axis=1))

            train_losses.append(train_loss)
            train_accuracies.append(train_acc)
            val_losses.append(val_loss)
            val_accuracies.append(val_acc)

            if epoch % 5 == 0 or epoch == epochs - 1:
                print(f"Epoch {epoch+1:3d}/{epochs} | "
                      f"Train Loss: {train_loss:.4f} Acc: {train_acc:.4f} | "
                      f"Val Loss: {val_loss:.4f} Acc: {val_acc:.4f}")

        return train_losses, train_accuracies, val_losses, val_accuracies

    def predict(self, X):
        return self.forward_propagation(X)

def build_tensorflow_model(input_shape=784):
    model = keras.Sequential([
        layers.Input(shape=(input_shape,)),
        layers.Dense(256, activation='relu', kernel_initializer='he_normal'),
        layers.BatchNormalization(),
        layers.Dropout(0.3),

        layers.Dense(128, activation='relu', kernel_initializer='he_normal'),
        layers.BatchNormalization(),
        layers.Dropout(0.2),

        layers.Dense(64, activation='relu', kernel_initializer='he_normal'),
        layers.Dropout(0.2),

        layers.Dense(10, activation='softmax')
    ], name='DigitRecognizer')

    model.compile(
        optimizer=keras.optimizers.Adam(learning_rate=0.001),
        loss='categorical_crossentropy',
        metrics=['accuracy']
    )

    return model

def load_and_preprocess_data():
    print("Loading MNIST dataset...")
    (x_train, y_train), (x_test, y_test) = keras.datasets.mnist.load_data()

    x_train = x_train.reshape(-1, 784).astype('float32') / 255.0
    x_test = x_test.reshape(-1, 784).astype('float32') / 255.0

    y_train_onehot = keras.utils.to_categorical(y_train, 10)
    y_test_onehot = keras.utils.to_categorical(y_test, 10)

    val_split = 5000
    x_val = x_train[-val_split:]
    y_val = y_train[-val_split:]
    y_val_onehot = y_train_onehot[-val_split:]

    x_train = x_train[:-val_split]
    y_train = y_train[:-val_split]
    y_train_onehot = y_train_onehot[:-val_split]

    print(f"Data loaded successfully!")
    print(f"   Training samples: {len(x_train):,}")
    print(f"   Validation samples: {len(x_val):,}")
    print(f"   Test samples: {len(x_test):,}")

    return (x_train, y_train, y_train_onehot,
            x_val, y_val, y_val_onehot,
            x_test, y_test, y_test_onehot)

def plot_sample_images(x_data, y_data, n_samples=10):
    fig, axes = plt.subplots(2, 5, figsize=(15, 6))
    fig.suptitle('Sample MNIST Images', fontsize=16, fontweight='bold')

    for i, ax in enumerate(axes.flat):
        ax.imshow(x_data[i].reshape(28, 28), cmap='gray')
        ax.set_title(f'Label: {y_data[i]}', fontsize=12)
        ax.axis('off')

    plt.tight_layout()
    plt.show()

def plot_training_history(history_dict):
    fig, axes = plt.subplots(1, 2, figsize=(15, 5))

    axes[0].plot(history_dict.get('train_loss', history_dict.get('loss')),
                 label='Training Loss', linewidth=2, color='#e74c3c')
    axes[0].plot(history_dict.get('val_loss'),
                 label='Validation Loss', linewidth=2, color='#3498db')
    axes[0].set_title('Model Loss', fontsize=14, fontweight='bold')
    axes[0].set_xlabel('Epoch', fontsize=12)
    axes[0].set_ylabel('Loss', fontsize=12)
    axes[0].legend(fontsize=10)
    axes[0].grid(True, alpha=0.3)

    axes[1].plot(history_dict.get('train_acc', history_dict.get('accuracy')),
                 label='Training Accuracy', linewidth=2, color='#2ecc71')
    axes[1].plot(history_dict.get('val_acc', history_dict.get('val_accuracy')),
                 label='Validation Accuracy', linewidth=2, color='#9b59b6')
    axes[1].set_title('Model Accuracy', fontsize=14, fontweight='bold')
    axes[1].set_xlabel('Epoch', fontsize=12)
    axes[1].set_ylabel('Accuracy', fontsize=12)
    axes[1].legend(fontsize=10)
    axes[1].grid(True, alpha=0.3)

    plt.tight_layout()
    plt.show()

def plot_confusion_matrix(y_true, y_pred):
    from sklearn.metrics import confusion_matrix
    import seaborn as sns

    cm = confusion_matrix(y_true, y_pred)

    plt.figure(figsize=(10, 8))
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues',
                xticklabels=range(10), yticklabels=range(10))
    plt.title('Confusion Matrix', fontsize=16, fontweight='bold')
    plt.xlabel('Predicted Label', fontsize=12)
    plt.ylabel('True Label', fontsize=12)
    plt.show()

def visualize_predictions(model, x_data, y_data, n_samples=10, model_type='tf'):
    indices = np.random.choice(len(x_data), n_samples, replace=False)

    fig, axes = plt.subplots(2, 5, figsize=(15, 6))
    fig.suptitle('Model Predictions', fontsize=16, fontweight='bold')

    for idx, ax in zip(indices, axes.flat):
        img = x_data[idx].reshape(1, 784)

        if model_type == 'tf':
            pred = model.predict(img, verbose=0)
        else:
            pred = model.predict(img)

        predicted_label = np.argmax(pred)
        confidence = np.max(pred) * 100
        true_label = y_data[idx]

        ax.imshow(x_data[idx].reshape(28, 28), cmap='gray')

        color = 'green' if predicted_label == true_label else 'red'
        ax.set_title(f'True: {true_label} | Pred: {predicted_label}\nConf: {confidence:.1f}%',
                    color=color, fontsize=10)
        ax.axis('off')

    plt.tight_layout()
    plt.show()

def plot_prediction_probabilities(model, x_data, idx, y_data, model_type='tf'):
    img = x_data[idx].reshape(1, 784)

    if model_type == 'tf':
        pred = model.predict(img, verbose=0)[0]
    else:
        pred = model.predict(img)[0]

    fig, axes = plt.subplots(1, 2, figsize=(12, 4))

    axes[0].imshow(x_data[idx].reshape(28, 28), cmap='gray')
    axes[0].set_title(f'True Label: {y_data[idx]}', fontsize=14, fontweight='bold')
    axes[0].axis('off')

    axes[1].bar(range(10), pred, color='steelblue', alpha=0.8)
    axes[1].set_xlabel('Digit', fontsize=12)
    axes[1].set_ylabel('Probability', fontsize=12)
    axes[1].set_title('Prediction Probabilities', fontsize=14, fontweight='bold')
    axes[1].set_xticks(range(10))
    axes[1].grid(True, alpha=0.3, axis='y')

    predicted_digit = np.argmax(pred)
    axes[1].bar(predicted_digit, pred[predicted_digit], color='green', alpha=0.8)

    plt.tight_layout()
    plt.show()

    print(f"Predicted Digit: {predicted_digit} (Confidence: {pred[predicted_digit]*100:.2f}%)")

print("\n" + "="*70)
print("VISION AI FUNDAMENTALS - DIGIT RECOGNIZER")
print("="*70 + "\n")

data = load_and_preprocess_data()
x_train, y_train, y_train_onehot = data[0], data[1], data[2]
x_val, y_val, y_val_onehot = data[3], data[4], data[5]
x_test, y_test, y_test_onehot = data[6], data[7], data[8]

print("\n" + "="*70)
print("DATASET EXPLORATION")
print("="*70 + "\n")

plot_sample_images(x_train, y_train)

plt.figure(figsize=(12, 4))
unique, counts = np.unique(y_train, return_counts=True)
plt.bar(unique, counts, color='steelblue', alpha=0.8)
plt.xlabel('Digit', fontsize=12)
plt.ylabel('Count', fontsize=12)
plt.title('Training Data Distribution', fontsize=14, fontweight='bold')
plt.xticks(range(10))
plt.grid(True, alpha=0.3, axis='y')
plt.show()

print("\n" + "="*70)
print("TRAINING NEURAL NETWORK FROM SCRATCH")
print("="*70 + "\n")

nn_scratch = NeuralNetworkScratch([784, 128, 64, 10])

train_losses, train_accs, val_losses, val_accs = nn_scratch.train(
    x_train[:30000], y_train_onehot[:30000],
    x_val, y_val_onehot,
    epochs=30,
    learning_rate=0.1,
    batch_size=128
)

print("\nTraining History:")
plot_training_history({
    'train_loss': train_losses,
    'train_acc': train_accs,
    'val_loss': val_losses,
    'val_acc': val_accs
})

test_pred = nn_scratch.predict(x_test)
test_acc = np.mean(np.argmax(test_pred, axis=1) == y_test)
print(f"\nFrom Scratch Model - Test Accuracy: {test_acc*100:.2f}%")

print("\nSample Predictions (From Scratch):")
visualize_predictions(nn_scratch, x_test, y_test, n_samples=10, model_type='scratch')

print("\n" + "="*70)
print("TRAINING TENSORFLOW MODEL")
print("="*70 + "\n")

tf_model = build_tensorflow_model()
print(tf_model.summary())

print("\nTraining TensorFlow model...")
history = tf_model.fit(
    x_train, y_train_onehot,
    epochs=20,
    batch_size=128,
    validation_data=(x_val, y_val_onehot),
    verbose=1
)

print("\nTraining History:")
plot_training_history(history.history)

test_loss, test_acc = tf_model.evaluate(x_test, y_test_onehot, verbose=0)
print(f"\nTensorFlow Model - Test Accuracy: {test_acc*100:.2f}%")

print("\nSample Predictions (TensorFlow):")
visualize_predictions(tf_model, x_test, y_test, n_samples=10, model_type='tf')

print("\n" + "="*70)
print("MODEL COMPARISON")
print("="*70 + "\n")

tf_pred = np.argmax(tf_model.predict(x_test, verbose=0), axis=1)
print("Confusion Matrix (TensorFlow Model):")
plot_confusion_matrix(y_test, tf_pred)

print("\n" + "="*70)
print("INTERACTIVE PREDICTION")
print("="*70 + "\n")

def predict_custom_image(uploaded_file, model, model_type='tf'):
    try:
        image = Image.open(io.BytesIO(uploaded_file['content'])).convert('L')
        image = ImageOps.invert(image)
        image = image.resize((28, 28))

        img_array = np.array(image).reshape(1, 784).astype('float32') / 255.0

        if model_type == 'tf':
            pred = model.predict(img_array, verbose=0)
        else:
            pred = model.predict(img_array)

        predicted_digit = np.argmax(pred)
        confidence = np.max(pred) * 100

        fig, axes = plt.subplots(1, 2, figsize=(12, 4))

        axes[0].imshow(np.array(image), cmap='gray')
        axes[0].set_title('Uploaded Image', fontsize=14, fontweight='bold')
        axes[0].axis('off')

        axes[1].bar(range(10), pred[0], color='steelblue', alpha=0.8)
        axes[1].bar(predicted_digit, pred[0][predicted_digit], color='green', alpha=0.8)
        axes[1].set_xlabel('Digit', fontsize=12)
        axes[1].set_ylabel('Probability', fontsize=12)
        axes[1].set_title(f'Prediction: {predicted_digit} ({confidence:.1f}%)',
                         fontsize=14, fontweight='bold')
        axes[1].set_xticks(range(10))
        axes[1].grid(True, alpha=0.3, axis='y')

        plt.tight_layout()
        plt.show()

        print(f"\nPredicted Digit: {predicted_digit}")
        print(f"   Confidence: {confidence:.2f}%")

    except Exception as e:
        print(f"Error processing image: {e}")

print("Upload a digit image to test the model:")
print("(The image should be a handwritten digit, ideally 28x28 pixels)")
print("\nRun the following cell to upload an image:")

print("""
uploaded = files.upload()
for filename, content in uploaded.items():
    predict_custom_image({'content': content}, tf_model, model_type='tf')
""")

print("\n" + "="*70)
print("INTERACTIVE TESTING")
print("="*70 + "\n")

print("Test with random samples from the test set:")
for i in range(3):
    idx = np.random.randint(0, len(x_test))
    print(f"\n--- Sample {i+1} ---")
    plot_prediction_probabilities(tf_model, x_test, idx, y_test, model_type='tf')

print("\n" + "="*70)
print("PROJECT COMPLETE!")
print("="*70)
print(f"\nSummary:")
print(f"   Neural Network from Scratch: {np.mean(np.argmax(nn_scratch.predict(x_test), axis=1) == y_test)*100:.2f}% accuracy")
print(f"   TensorFlow Model: {test_acc*100:.2f}% accuracy")
print(f"   Both models achieve 95%+ accuracy on MNIST!")
print("\nYou can now:")
print("   1. Upload your own digit images for prediction")
print("   2. Experiment with different architectures")
print("   3. Tune hyperparameters for better performance")
print("   4. Export models for deployment")
print("="*70)
```
